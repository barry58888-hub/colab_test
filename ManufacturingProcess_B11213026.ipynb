```python
import barcode
from barcode.writer import ImageWriter
from IPython.display import Image, display

#產品料號
product_id = input ("林建均製程條碼展示-請輸入你設計的物料編號")

# ［很重要］建立 code128條碼 （未來實務專題可以把code128改成廠商類型）
code = barcode.get('code128',product_id, writer=ImageWriter())

#儲存條碼
filename = code.save("product_barcode")

#顯示條碼
display(Image(filename=filename))
```

```python
#安裝套件
import qrcode
from IPython.display import display
manufacturing_info ="""
Product: YUNPID-2026-001
Batch: B24001
Machine: CNC-03
Date: 2027-05-07
Operator: Yuhsin Hung
"""
#生成二維條碼
qr = qrcode.make(manufacturing_info)
#儲存二維條碼
qr.save("manufacturing_qrcode.png")
#顯示二維條碼
display(qr)
```
